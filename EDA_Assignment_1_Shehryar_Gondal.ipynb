__Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.__

The Wine Quality dataset contains 12 features, including 11 input variables and 1 output variable (quality). The input variables are:

* fixed acidity: the amount of acid in the wine, which affects its taste and texture.
* volatile acidity: the amount of acetic acid in the wine, which can contribute to a sour taste and unpleasant smell.
* citric acid: the amount of citric acid in the wine, which can contribute to a fresh and fruity taste.
* residual sugar: the amount of sugar left in the wine after fermentation, which can affect the sweetness and body of the wine.
* chlorides: the amount of salt in the wine, which can contribute to a salty or bitter taste.
* free sulfur dioxide: the amount of sulfur dioxide that is free in the wine, which can prevent oxidation and microbial spoilage.
* total sulfur dioxide: the total amount of sulfur dioxide in the wine, which can also prevent oxidation and microbial spoilage.
* density: the mass of the wine per unit volume, which can be used to determine the alcohol content and sweetness of the wine.
* pH: the level of acidity or alkalinity in the wine, which can affect its taste, color, and stability.
* sulphates: the amount of sulfur compounds in the wine, which can act as an antioxidant and contribute to the wine's flavor and aroma.
* alcohol: the percentage of alcohol in the wine, which can affect its taste, texture, and aroma.

__The output variable, quality, is a score between 0 and 10 that rates the overall quality of the wine.__

__Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?__

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Wine Quality dataset
wine = pd.read_csv('D:\JupyterNotebook\PW_Skills_Data_Science\Exploratory Data Analysis (EDA)\Hindi\winequality-red.csv')

# Plot histograms
sns.set(style="ticks", color_codes=True)
for column in wine.columns:
    sns.displot(wine[column], kde=True)
    plt.title(column)
    plt.show()
```

__Based on the histograms, it appears that several features exhibit non-normality, including fixed acidity, volatile acidity, citric acid, residual sugar, and chlorides. To improve normality, we can apply transformations such as log transformation, square root transformation, or Box-Cox transformation. For example, we can apply a log transformation to the residual sugar feature to make it more normally distributed:__

```python
import numpy as np

# Log transform residual sugar feature
wine['residual sugar'] = np.log(wine['residual sugar'])

# Plot histogram of transformed
sns.displot(wine['residual sugar'], kde=True)
plt.title('Log Transformed Residual Sugar')
plt.show()
```

__Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?__

```python
import pandas as pd
from sklearn.decomposition import PCA

# Load Wine Quality dataset
wine = pd.read_csv('D:\JupyterNotebook\PW_Skills_Data_Science\Exploratory Data Analysis (EDA)\Hindi\winequality-red.csv')

# Separate features and target variable
X = wine.drop('quality', axis=1)
y = wine['quality']

# Fit PCA with all features
pca = PCA().fit(X)

# Calculate cumulative explained variance ratio
variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = [sum(variance_ratio[:i+1]) for i in range(len(variance_ratio))]

# Determine minimum number of principal components required to explain 90% of variance
min_components = sum([cumulative_variance_ratio[i] < 0.9 for i in range(len(cumulative_variance_ratio))]) + 1
print("Minimum number of principal components required to explain 90% of variance:", min_components)
```
